```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
import warnings; warnings.filterwarnings('ignore')

#Train models libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


#Evaluation librarties
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

#Save models
from joblib import dump
from joblib import load
```

```python
#Load Models

#Balance Models
lr_model_b = load('Models/Logistic_Regression_Balance.joblib')
dt_model_b = load('Models/Decision_Tree_Balance.joblib')
rf_model_b = load('Models/Random_Forest_Balance.joblib')
xgb_model_b = load('Models/XGBoost_Balance.joblib')

#Imbalanced Models
lr_model_i = load('Models/Logistic_Regression_Imbalance.joblib')
dt_model_i = load('Models/Decision_Tree_Imbalance.joblib')
rf_model_i = load('Models/Random_Forest_Imbalance.joblib')
xgb_model_i = load('Models/XGBoost_Imbalance.joblib')
```

```python
#Get data

file_path = 'Data/cleaned_combined_user_profiles.csv'
df = pd.read_csv(file_path)

# Step 2: Display shape and first few rows
print("Dataset shape:", df.shape)
display(df.head())
```

### Split Train and Test Imbalance

```python
features = ['SOC', 'SOH', 'Charging_Cycles', 'Battery_Temp', 'Motor_RPM',
            'Motor_Torque', 'Motor_Temp', 'Brake_Pad_Wear', 'Charging_Voltage', 'Tire_Pressure']

X = df[features]
y = df['DTC']

# Confirm the dimensions of the input and output
print("Shape of X (features):", X.shape)
print("Shape of y (target):", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Confirm shapes
print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)

# Check class distribution in training and test sets
print("\nTraining set class distribution:")
print(y_train.value_counts(normalize=True) * 100)

print("\nTest set class distribution:")
print(y_test.value_counts(normalize=True) * 100)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Confirm shapes
print("Scaled training features shape:", X_train_scaled.shape)
print("Scaled test features shape:", X_test_scaled.shape)
```

### Get Predictions Imbalance

```python
y_pred_lr_i = lr_model_i.predict(X_test_scaled)
y_proba_lr_i = lr_model_i.predict_proba(X_test_scaled)[:, 1] 

y_pred_dt_i = dt_model_i.predict(X_test)
y_proba_dt_i = dt_model_i.predict_proba(X_test)[:, 1] 

y_pred_rf_i = rf_model_i.predict(X_test)
y_proba_rf_i = rf_model_i.predict_proba(X_test)[:, 1] 

y_pred_xgb_i = xgb_model_i.predict(X_test)
y_proba_xgb_i = xgb_model_i.predict_proba(X_test)[:, 1] 
```

### Split Train and Test Balance

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check new class distribution
print("Balanced training set shape:", X_train_smote.shape, y_train_smote.shape)
print("\nClass distribution after SMOTE:")
print(y_train_smote.value_counts())
print("\nPercentage:")
print(y_train_smote.value_counts(normalize=True) * 100)
```

```python
# Step 9: Scale features for Logistic Regression

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)  # Apply to SMOTE data
X_test_scaled = scaler.transform(X_test)              # Test set remains unchanged

# Confirm shapes
print("Scaled training features shape:", X_train_scaled.shape)
print("Scaled test features shape:", X_test_scaled.shape)
```

### Get Predictions Balance

```python
y_pred_lr_b = lr_model_b.predict(X_test_scaled)
y_proba_lr_b = lr_model_b.predict_proba(X_test_scaled)[:, 1] 

y_pred_dt_b = dt_model_b.predict(X_test)
y_proba_dt_b = dt_model_b.predict_proba(X_test)[:, 1] 

y_pred_rf_b = rf_model_b.predict(X_test)
y_proba_rf_b = rf_model_b.predict_proba(X_test)[:, 1] 

y_pred_xgb_b = xgb_model_b.predict(X_test)
y_proba_xgb_b = xgb_model_b.predict_proba(X_test)[:, 1] 
```

### Get Results

```python
results = pd.DataFrame({
    'Model': ['Logistic Regression (Imb)', 'Logistic Regression (Bal)',
              'Decision Tree (Imb)', 'Decision Tree (Bal)',
              'Random Forest (Imb)', 'Random Forest (Bal)',
              'XGBoost (Imb)', 'XGBoost (Bal)'],
    'Accuracy': [
        accuracy_score(y_test, y_pred_lr_i),
        accuracy_score(y_test, y_pred_lr_b),
        accuracy_score(y_test, y_pred_dt_i),
        accuracy_score(y_test, y_pred_dt_b),
        accuracy_score(y_test, y_pred_rf_i),
        accuracy_score(y_test, y_pred_rf_b),
        accuracy_score(y_test, y_pred_xgb_i),
        accuracy_score(y_test, y_pred_xgb_b)
    ],
    'Balanced Accuracy': [
        balanced_accuracy_score(y_test, y_pred_lr_i),
        balanced_accuracy_score(y_test, y_pred_lr_b),
        balanced_accuracy_score(y_test, y_pred_dt_i),
        balanced_accuracy_score(y_test, y_pred_dt_b),
        balanced_accuracy_score(y_test, y_pred_rf_i),
        balanced_accuracy_score(y_test, y_pred_rf_b),
        balanced_accuracy_score(y_test, y_pred_xgb_i),
        balanced_accuracy_score(y_test, y_pred_xgb_b)
    ],
    'Precision': [
        precision_score(y_test, y_pred_lr_i),
        precision_score(y_test, y_pred_lr_b),
        precision_score(y_test, y_pred_dt_i),
        precision_score(y_test, y_pred_dt_b),
        precision_score(y_test, y_pred_rf_i),
        precision_score(y_test, y_pred_rf_b),
        precision_score(y_test, y_pred_xgb_i),
        precision_score(y_test, y_pred_xgb_b)
    ],
    'Recall': [
        recall_score(y_test, y_pred_lr_i),
        recall_score(y_test, y_pred_lr_b),
        recall_score(y_test, y_pred_dt_i),
        recall_score(y_test, y_pred_dt_b),
        recall_score(y_test, y_pred_rf_i),
        recall_score(y_test, y_pred_rf_b),
        recall_score(y_test, y_pred_xgb_i),
        recall_score(y_test, y_pred_xgb_b)
    ],
    'F1 Score': [
        f1_score(y_test, y_pred_lr_i),
        f1_score(y_test, y_pred_lr_b),
        f1_score(y_test, y_pred_dt_i),
        f1_score(y_test, y_pred_dt_b),
        f1_score(y_test, y_pred_rf_i),
        f1_score(y_test, y_pred_rf_b),
        f1_score(y_test, y_pred_xgb_i),
        f1_score(y_test, y_pred_xgb_b)
    ],
    'ROC AUC': [
        roc_auc_score(y_test, y_proba_lr_i),
        roc_auc_score(y_test, y_proba_lr_b),
        roc_auc_score(y_test, y_proba_dt_i),
        roc_auc_score(y_test, y_proba_dt_b),
        roc_auc_score(y_test, y_proba_rf_i),
        roc_auc_score(y_test, y_proba_rf_b),
        roc_auc_score(y_test, y_proba_xgb_i),
        roc_auc_score(y_test, y_proba_xgb_b)
    ]
})

# Round for readability
results_rounded = results.round(4)

# Display the table
results_rounded
```

```python
results['Type'] = results['Model'].str.extract(r'\((Imb|Bal)\)')

df_melted = pd.melt(
    results,
    id_vars=['Model', 'Type'],
    value_vars=['F1 Score', 'Balanced Accuracy', 'Recall'],
    var_name='Metric',
    value_name='Score'
)

plt.figure(figsize=(14, 6))
ax = sns.barplot(
    data=df_melted,
    x='Model',
    y='Score',
    hue='Metric',
    palette='viridis',
    alpha=0.8,
    dodge=True
)

for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2., 
        height + 0.01,                  
        f'{height:.2f}',                   
        ha='center',                 
        va='bottom',                   
        fontsize=10
    )

plt.title('Evaluation Metrics Comparisson (Imb y Bal)', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1.15) 
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
results['Type'] = ['Imb', 'Bal', 'Imb', 'Bal', 'Imb', 'Bal', 'Imb', 'Bal']
results['Model'] = ['LR', 'LR', 'DT', 'DT', 'RF', 'RF', 'XGB', 'XGB']

plt.figure(figsize=(10, 6))
ax = sns.barplot(
    data=results,
    x='Model',
    y='ROC AUC',
    hue='Type',
    palette=['#FF6B6B', '#4ECDC4']
)

# Añadir valores
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.3f}',
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

plt.title('ROC AUC COMPARISSON')
plt.legend(title='Tipo')
plt.tight_layout()
plt.show()
```
